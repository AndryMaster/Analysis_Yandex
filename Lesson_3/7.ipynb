```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


def pep8_columns(dataframe: pd.DataFrame) -> pd.DataFrame:
    dataframe.columns = [col.lower().replace(' ', '_') for col in dataframe.columns]
    return dataframe


df = pd.read_csv('data_1.csv', sep=',', encoding='cp1251')  # encoding='' utf-8 cp1251
df = pep8_columns(df).dropna()

y_param = input()  # 'ингредиент_1'

plt.title('Диаграмма рассеяния:\nАнализ зависимости между факторами', size=16, y=1.01)
sns.scatterplot(data=df, x='вес', y=y_param, hue='оператор_линии')

plt.savefig('target_3_7.png')  # plt.show()
```
