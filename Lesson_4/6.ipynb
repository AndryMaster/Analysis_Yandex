```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


def pep8_columns(dataframe: pd.DataFrame) -> pd.DataFrame:
    dataframe.columns = [col.lower().replace(' ', '_') for col in dataframe.columns]
    return dataframe


df = pd.read_csv('6.csv', sep=';')
# df = pd.read_csv('data.csv', sep=',')
df = pep8_columns(df).dropna()
# покупательская_активность, выручка_от_клиента_текущий_месяц

# y_param = input()
y_param = 'Полёт'
data = df[df['покупательская_активность'] == y_param]
data.rename({'выручка_от_клиента_текущий_месяц': '0'}, axis='columns', inplace=True)


fig, axs = plt.subplots(ncols=2, figsize=(9, 4))  # fig.set_size_inches(9, 4)

plt.suptitle('Гистограмма и ящик с усами для количественных данных')

g1 = sns.histplot(data=data, bins=30, kde=True, legend=False, ax=axs[0])
g2 = sns.boxplot(data=data, ax=axs[1])

g1.set(ylabel=None)
g2.set(xlabel=None)

plt.savefig('target_4_6.png')  # plt.show()
```
