Продолжаем работать с данными туристической компании.

1) Загрузить данные из файла. Разделителем между значениями является запятая.

2) На основе количественного фактора расстояние создать новый категориальный фактор расстояние_кат.

Всё, что меньше равно 300 км назвать домашний_регион
Всё, что более 300 км, но менее равно 700 км назвать недалеко_отдома
Всё, что более 700 км, назвать дальнее_путешествие
3) С помощью изученной в теории подхода, посчитать долю клиентов по категориям домашний_регион, недалеко_отдома и дальнее_путешествие из созданного в п.2 фактора расстояние_кат. Результат вывести в процентах и округлить до целого по математическому правилу.

```python
import pandas as pd

# df = pd.read_csv('data.csv', sep=',', skipinitialspace=True)

df = pd.read_csv(PATH_OUT + '\\tema_13_teor.csv', sep=';', skipinitialspace=True, encoding='cp1251')
df.columns = df.columns.str.lower().str.replace(' ', '_')
df['возраст'] = df['возраст'].str.replace(',', '.')
df['возраст'] = df['возраст'].astype(float)

df['расстояние_кат'] = df['расстояние'].apply(lambda x: 'домашний_регион' if x<= 300 \
                        else ('не_так_далеко' if x<= 700 else 'дальнее_путешествие'))

print(round(df[['расстояние_кат']].value_counts(normalize=True)*100, 0).astype(int))
```

## Задача 6

Вы будете работать с факторами:
- общая_оценка_качества_предоставленной_услуги. Имеет категории: 'плохо', 'нормально', 'хорошо'
- путешествует_с_детьми. Имеет категории: 'да', 'нет'
- расстояние_кат. Имеет категории: 'дальнее_путешествие', 'домашний_регион', 'не_так_далеко'

1) Загрузить данные из этого файла. Разделителем между значениями является запятая.

2) Посчитать долю плохих оценок качества обслуживания общая_оценка_качества_предоставленной_услуги, исходя из типа путешествия “расстояние_кат” и наличия или отсутствия детей “путешествует_с_детьми”. На экран вывести результат в процентах с округлением до одного знака после запятой.
