```python
from scipy.stats import chi2_contingency
import pandas as pd


df = pd.read_csv("data3.csv", sep=';', decimal=',')

result = []
for col in df.columns[2:]:
    contingency_table = pd.crosstab(df["вид_зондирования"], df[col])
    res = chi2_contingency(contingency_table)
    result.append([col, round(float(res.pvalue), 4)])

print([['показатель', 'р-уровень']] + sorted(result, key=lambda x: x[1]))
```
