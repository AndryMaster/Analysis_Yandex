```python
from scipy.stats import shapiro, spearmanr, pearsonr
import pandas as pd


def r(x, n):
    return float(round(x, n))


df = pd.read_csv("data4.csv", sep=';', decimal=',')

# print(df.info())
# print(df.describe())
df = df[df["Группа"] == "Группа_2"]

cols = df.columns[1:]
result = []

for i in range(len(cols)):
    for j in range(i + 1, len(cols)):
        g1, g2 = df[cols[i]], df[cols[j]]

        row = [cols[i], cols[j]]
        if shapiro(g1).pvalue >= 0.05 and shapiro(g2).pvalue >= 0.05:
            res = pearsonr(g1, g2)
            row += [r(res.statistic, 2), r(res.pvalue, 4), "Пирсон"]
        else:
            res = spearmanr(g1, g2)
            row += [r(res.statistic, 2), r(res.pvalue, 4), "Спирмен"]

        if res.pvalue < 0.05:
            result.append(row)

print([['показатель 1', 'показатель 2', 'значение корреляции', 'р-уровень', 'метод корреляции']]
      + sorted(result, key=lambda x: x[2], reverse=True))
```
