```python
from scipy import stats
import pandas as pd


df = pd.read_csv("data1.csv", sep=';')
regions = df["регион"].unique()

result = []
for region in regions:
    sub_df = df[df["регион"] == region]
    res = stats.mannwhitneyu(sub_df["время_до"], sub_df["время_после"])
    # pearsonr spearmanr kendalltau wilcoxon ttest_ind median_test
    result.append([region, round(float(res.pvalue), 4)])

print([['регион', 'р-уровень']] + sorted(result, key=lambda x: x[-1]))
# [['регион', 'р-уровень'], ['Центр', 0.004], ['Урал', 0.0094], ['Северо-запад', 0.5067]]
```
